```python
import requests
url = "https://product.kyobobook.co.kr/api/gw/pub/pdt/best-seller/total?page=1&per=20&period=002&bsslBksClstCode=A"
head = {'User-Agent':
'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'}
r = requests.get(url, headers=head)
```

```python
r.text
```

```python
r.json()['data']['bestSeller']
```

```python
r.json()['data']['bestSeller']['productInfo']
```

```python
import pandas as pd
df = pd.DataFrame(r.json()['data']['bestSeller'])
```

```python
pd.DataFrame([df['productInfo'].iloc[0]])
```

```python
data = pd.concat([df, pd.DataFrame(df['productInfo'].apply(lambda x : x).tolist())], axis=1)
```

```python
data.drop(['productInfo'], axis=1, inplace=True)
```

```python
data.to_csv("./aaa.csv", index=False)
```
